```python
from sklearn.model_selection import train_test_split
```

```python
input_data = data.drop(columns=['loan_status'])
output_data = data['loan_status']
```

```python
x_train,x_test,y_train,y_test = train_test_split(input_data,output_data, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
```

```python
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

```python
import gradio as gr
import pandas as pd
import pickle as pk

model = pk.load(open("model.pkl", "rb"))
scaler = pk.load(open("scaler.pkl", "rb"))

# Define prediction function
def loan_prediction(no_of_dep, grad, self_emp, Annual_Income, Loan_Amount, Loan_Dur, Cibil, Assets):
    # Convert categorical inputs
    grad_s = 0 if grad == "Graduated" else 1
    emp_s = 0 if self_emp == "No" else 1

    # Prepare data
    pred_data = pd.DataFrame([[no_of_dep, grad_s, emp_s, Annual_Income, Loan_Amount, Loan_Dur, Cibil, Assets]],
                             columns=['no_of_dependents','education','self_employed','income_annum','loan_amount','loan_term','cibil_score','Assets'])
    # Scale and predict
    pred_data = scaler.transform(pred_data)
    prediction = model.predict(pred_data)
    return "Loan Approved" if prediction[0] == 1 else "Loan Rejected"

#Gradio interface
demo = gr.Interface(
    fn=loan_prediction,
    inputs=[
        gr.Slider(0, 5, step=1, label="Number of Dependents"),
        gr.Radio(["Graduated", "Not Graduated"], label="Education"),
        gr.Radio(["Yes", "No"], label="Self Employed"),
        gr.Slider(0, 10000000, step=10000, label="Annual Income"),
        gr.Slider(0, 10000000, step=10000, label="Loan Amount"),
        gr.Slider(0, 20, step=1, label="Loan Duration (Years)"),
        gr.Slider(0, 1000, step=10, label="Cibil Score"),
        gr.Slider(0, 10000000, step=10000, label="Assets")
    ],
    outputs=gr.Textbox(label="Prediction Result"),
    title="Loan Prediction App",
    description="Predict loan approval based on user details using a trained machine learning model."
)

demo.launch()

```
